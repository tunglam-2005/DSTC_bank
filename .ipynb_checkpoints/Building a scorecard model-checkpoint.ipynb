```python
X_1 = data.drop(columns=['customer_id','label'])
y_1 = data['label']
X_1.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_1, y_1, test_size=0.3, random_state=42)
model = AdaBoostClassifier(algorithm='SAMME')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

tn, fp, fn, tp = metrics.confusion_matrix(y_test, y_pred).ravel()
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred)
recall = metrics.recall_score(y_test, y_pred)

print(f"True positives: {tp}")
print(f"False positives: {fp}")
print(f"True negatives: {tn}")
print(f"False negatives: {fn}\n")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```

```python
binned_data = pd.DataFrame()
for col in X.columns:
    binned_data[col] = pd.cut(data[col], bins=5)
X_train_binned, X_test_binned, y_train, y_test = train_test_split(binned_data, y, test_size=0.3, random_state=42)
```

```python
base_score = 600
pdo = 20
factor = pdo / np.log(2)
offset = base_score - (factor * np.log(50))
```

```python
def calculate_adaboost_score_by_group_with_weights(model, base_score, X_binned, X_original):
    factor = pdo / np.log(2)
    offset = base_score - factor * np.log(50)
    estimator_weights = model.estimator_weights_
    group_scores = {feature: {group: 0 for group in X_binned[feature].unique()} for feature in X_binned.columns}
    estimator_contributions = {feature: 0 for feature in X_binned.columns}
    for i in range(X_binned.shape[0]):
        for feature in X_binned.columns:
            sample_group = X_binned.iloc[i][feature]  
            original_data_point = X_original.iloc[i].values.reshape(1, -1)  
            sample_score = offset
            for j, estimator in enumerate(model.estimators_):
                weak_pred = estimator.predict(original_data_point)[0]  
                sample_score += factor * estimator_weights[j] * weak_pred  
                estimator_contributions[feature] += estimator_weights[j]
            group_scores[feature][sample_group] += sample_score
    for feature, group_scores_dict in group_scores.items():
        print(f"Feature: {feature}")
        for group, score in group_scores_dict.items():
            print(f"  Nhóm {group}: {score:.2f}")
        print(f"  Tổng trọng số ước lượng cho {feature}: {estimator_contributions[feature]:.4f}\n")
    return group_scores
customer_group_scores = calculate_adaboost_score_by_group_with_weights(model, base_score, X_test_binned, X)
```

```python
def calculate_adaboost_score(model, base_score, pdo, X):
    factor = pdo / np.log(2)
    offset = base_score - factor * np.log(50)
    estimator_weights = model.estimator_weights_
    scores = []
    for i in range(X.shape[0]):
        sample_score = offset
        for j, estimator in enumerate(model.estimators_):
            weak_pred = estimator.predict([X.iloc[i]])[0]
            sample_score += factor * estimator_weights[j] * weak_pred
        scores.append(sample_score)
    return scores

customer_scores = calculate_adaboost_score(model, base_score, pdo, X_test)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

youden_j = tpr - fpr
optimal_idx = np.argmax(youden_j)
best_threshold = thresholds[optimal_idx]

def probability_to_score(prob, base_score, pdo):
    factor = pdo / np.log(2)
    odds = (1 - prob) / prob
    score = base_score + factor * np.log(odds)
    return score

cutoff_probability = best_threshold
cutoff_score = probability_to_score(cutoff_probability, base_score, pdo)

print(f"Ngưỡng xác suất cutoff tối ưu nhất: {cutoff_probability:.2f}")
print(f"Số điểm tương đương: {cutoff_score:.2f}")
```

```python
approved_customers = []
rejected_customers = []

for score in customer_scores:
    if score >= cutoff_score:
        approved_customers.append(score)
    else:
        rejected_customers.append(score)

print(f"Số khách hàng được chấp nhận hoàn toàn: {len(approved_customers)}")
print(f"Số khách hàng có thể bị bị từ chối: {len(rejected_customers)}")
```

```python
valid_thresholds = thresholds[(thresholds > 0) & (thresholds < 1)]
valid_youden_j = tpr[:len(valid_thresholds)] - fpr[:len(valid_thresholds)]
least_optimal_idx = np.argmin(valid_youden_j)
worst_threshold = valid_thresholds[least_optimal_idx]
worst_cutoff_score = probability_to_score(worst_threshold, base_score, pdo)
print(f"Ngưỡng xác suất cutoff kém tối ưu nhất: {worst_threshold:.2f}")
print(f"Số điểm tương đương: {worst_cutoff_score:.2f}")
```

```python
def classify_risk(score, low_risk_cutoff=cutoff_score, high_risk_cutoff=worst_cutoff_score):
    if score >= cutoff_score:
        return 'Rủi ro thấp'
    elif score > high_risk_cutoff:
        return 'Rủi ro trung bình'
    else:
        return 'Rủi ro cao'
customer_risk_segments = [classify_risk(score) for score in customer_scores]
risk_counts = Counter(customer_risk_segments)
for i, (score, risk) in enumerate(zip(customer_scores, customer_risk_segments)):
    print(f"Khách hàng {i+1}: Điểm = {score}, Phân khúc rủi ro = {risk}")
print("Số khách hàng ở mỗi phân khúc rủi ro thuộc tập test:")
for risk, count in risk_counts.items():
    print(f"{risk}: {count} khách hàng")
```
